```python
import pandas as pd


```

```python

# Load the dataset
file_path = 'ProjectFriendsdataset.csv'  
data = pd.read_csv(file_path)

```

```python

# Preview the first few rows
print(data.head())


```

```python


# Check for missing values
print("\nMissing values per column:")
print(data.isnull().sum())

```

```python
print("\nMissing values per column:")
print(data.isnull().sum())
```

```python
gender_counts = data['Gender'].value_counts()
gender_percentage = data['Gender'].value_counts(normalize=True) * 100
```

```python
print("Dialogue Counts by Gender:\n", gender_counts)
print("\nDialogue Percentage by Gender:\n", gender_percentage)
```

```python
main_characters_data = data[data['MainCharacter'] != 'Others']

# Calculate dialogue counts for each main character
dialogue_counts_by_main_character = main_characters_data['MainCharacter'].value_counts()
```

```python
print("Dialogue Counts by Main Character:\n", dialogue_counts_by_main_character)
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download the VADER lexicon if not already downloaded
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()

# Compute sentiment scores
data['Sentiment'] = data['Dialogue'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Categorize sentiment into positive, negative, or neutral
data['SentimentCategory'] = data['Sentiment'].apply(lambda x: 'Positive' if x > 0.05 else 'Negative' if x < -0.05 else 'Neutral')
```

```python
sentiment_by_gender = data.groupby('Gender')['SentimentCategory'].value_counts(normalize=True).unstack()
print("\nSentiment Distribution by Gender:\n", sentiment_by_gender)
```

```python
import matplotlib.pyplot as plt

# Gender-wise sentiment distribution
sentiment_by_gender.plot(kind='bar', figsize=(10, 6))
plt.title("Sentiment Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Proportion")
plt.xticks(rotation=0)
plt.show()
```

```python
# Filter the dataset for main characters only
main_characters_data = data[data['MainCharacter'] != 'Others']

# Calculate sentiment distribution by main character
sentiment_by_main_character = (
    main_characters_data.groupby(['MainCharacter', 'SentimentCategory'])
    .size()
    .unstack(fill_value=0)
)
# Normalize to get proportions
sentiment_by_main_character = sentiment_by_main_character.div(sentiment_by_main_character.sum(axis=1), axis=0)
```

```python
# Display the sentiment distribution
print("Sentiment Distribution by Main Character:\n", sentiment_by_main_character)

# Save the output to a CSV file (optional)
sentiment_by_main_character.to_csv('sentiment_distribution_by_main_character.csv')
```

```python
sentiment_by_main_character.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# Add titles and labels
plt.title("Sentiment Distribution by Main Character", fontsize=16)
plt.xlabel("Main Character", fontsize=12)
plt.ylabel("Proportion", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Sentiment Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group by Episode and SentimentCategory to see trends over time
sentiment_trends = data.groupby(['Episode', 'SentimentCategory']).size().unstack()

# Normalize the values to show proportions
sentiment_trends = sentiment_trends.div(sentiment_trends.sum(axis=1), axis=0)

# Plot the trend
sentiment_trends.plot(figsize=(12,6), title="Sentiment Trend Over Episodes")
plt.xlabel("Episode")
plt.ylabel("Proportion")
plt.legend(title="Sentiment Category")
plt.show()
```
